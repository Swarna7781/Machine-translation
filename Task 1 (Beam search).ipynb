# Importing necessary libraries

```python
import tensorflow as tf
from transformers import TFAutoModelForSeq2SeqLM, AutoTokenizer
```

# Using pre-trained model for translation

```python
model_name = "Helsinki-NLP/opus-mt-en-de"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = TFAutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
input_text = "This model translates english to german"
```

# Tokenizing the input sentence

```python
input_ids = tokenizer(input_text, return_tensors="tf").input_ids
```

```python
beam_size = 3
max_length = 50
```

# Applying beam search

```python
def beam_search_decoder(model, input_ids, beam_size, max_length):
    decoded_outputs = model.generate(
        input_ids=input_ids,
        max_length=max_length,
        num_beams=beam_size,
        early_stopping=True,
        num_return_sequences=beam_size,
        pad_token_id=tokenizer.pad_token_id,
        eos_token_id=tokenizer.eos_token_id,
    )
    return decoded_outputs
```

```python
translated_texts = beam_search_decoder(model, input_ids, beam_size, max_length)
```

```python
for i, output_ids in enumerate(translated_texts):
    decoded_text = tokenizer.decode(output_ids, skip_special_tokens=True)
    print(f"Translation {i+1}: {decoded_text}")
```
